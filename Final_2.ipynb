```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests

url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
data  = requests.get(url).text
soup = BeautifulSoup(data, 'html5lib')

#enlaces = soup.find('table').find_all('td')
postales = pd.DataFrame(columns=["CodigoPostal","Vecindario"])

for row in soup.find('table').find_all('td'):
  dato = row.find('b').text
  cont = row.find('span').text
  postales = postales.append({"CodigoPostal":dato,"Vecindario":cont}, ignore_index=True)

postales
```
